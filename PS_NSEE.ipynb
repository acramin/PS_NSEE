# Estruturação do banco de dados

## Adicionando e/ou alterando colunas

### Adicionando as colunas CONSDIAG, DIAGTRAT e TRATCONS

CONSDIAG – diferença em dias entre Diagnóstico e Consulta – 0 = até 30 dias; 1 = entre 31 e 60 dias; 2 = mais de 61 dias;

DIAGTRAT – diferença em dias entre Diagnóstico e Tratamento – 0 = até 60 dias; 1 = entre 61 e 90 dias; 2 = mais de 91 dias; 3 = Não tratou (datas de tratamento vazias);

TRATCONS – diferença em dias entre Tratamento e Consulta – 0 = até 60 dias; 1 = entre 61 e 90 dias; 2 = mais de 91 dias; 3 = Não tratou (datas de tratamento vazias).

```python
df_colo = df_idade.copy() ## alteração será feita nas colunas, então mudei o nome para fazer sentido pra mim :P

# Regra de ConsDiag
def regra_consdiag(dias):
    if dias <= 30:
        return 0
    elif 31 <= dias <= 60:
        return 1
    else:
        return 2

df_colo['CONSDIAG'] = df_colo['CONSDIAG'].apply(regra_consdiag) # aplica a regra nas linhas

# # Regra de DIAGTRAT
def regra_diagtrat(dias):
    if pd.isna(dias):  # Se a data de tratamento for vazia (NaT)
        return 3
    elif dias <= 60:
        return 0
    elif 61 <= dias <= 90:
        return 1
    else:
        return 2

df_colo['DIAGTRAT'] = df_colo['DIAGTRAT'].apply(regra_diagtrat) # aplica a regra nas linhas

# Regra de TRATCONS
def regra_tratcons(dias):
    if pd.isna(dias):  # Se a data de tratamento for vazia (NaT)
        return 3
    elif dias <= 60:
        return 0
    elif 61 <= dias <= 90:
        return 1
    else:
        return 2

df_colo['TRATCONS'] = df_colo['TRATCONS'].apply(regra_tratcons) # aplica a regra nas linhas

df_colo.DIAGTRAT.value_counts()
```

### Extraindo números das colunas DRS e DRSINST

```python
df_colo['DRS'] = df_colo['DRS'].str.extract('(\d+)')
df_colo.DRS.value_counts()
```

```python
df_colo['DRSINST'] = df_colo['DRSINST'].str.extract('(\d+)')
df_colo.DRSINST.value_counts()
```

### Criando coluna binária para ULTINFO

OBITO

0 - Morto

1 - Vivo

```python
def condicao_binaria(valor):
    if valor in [1, 2]:
        return 1  # Vivo
    elif valor in [3, 4]:
        return 0  # Óbito


df_colo['OBITO'] = df_colo['ULTINFO'].apply(condicao_binaria)
df_colo['OBITO'].value_counts()
```

# Pré-Processamento

## Informações Básicas

```python
from sklearn.model_selection import train_test_split
df = pd.read_csv("/content/pacigeral_jun24_limpo.csv")
print(df.info())
```

## Separando variáveis

```python
X = df.drop('OBITO', axis=1)
y = df['OBITO'] # saída
```

## Tratando variáveis categóricas



```python
from sklearn.preprocessing import LabelEncoder # feito dessa forma para manter a quantidade de colunas - testei com get_dummies mas de 35 col foi para 128
variaveis_categoricas = X.select_dtypes(include=['object']).columns  # Seleciona as colunas categóricas
print(variaveis_categoricas)
```

```python
encoders = {}  # armazena relação original : código

for col in variaveis_categoricas:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    encoders[col] = le

for col, encoder in encoders.items(): # exibe relação original : código
    print(f'Classes para {col}:')
    print(dict(zip(encoder.classes_, encoder.transform(encoder.classes_))))
```

## Dividindo dados entre treino e teste

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) # divide dados em 4 conjuntos

# Verificar as dimensões - relação 80% train / 20% test
print("Tamanho do conjunto de treinamento:", X_train.shape[0])
print("Tamanho do conjunto de teste:", X_test.shape[0])

X_train.head()
```
